## display different colors for each species

```python
# create an output file called penguins.html
# output_file("output-files/penguins.html")

# create a figure object
fig = figure()

# tools
fig.tools = [PanTool(), ResetTool()]
fig.add_tools(HoverTool())
fig.toolbar.logo = None

fig.width = 1000
fig.height = 800
fig.background_fill_color = "azure"

# style the title
fig.title.text = "Penguins mass vs flipper length (in mm)"
fig.title.text_font = "arial"
fig.title.text_font_size = "24px"
fig.title.align = "center"

# style the axes
fig.axis.major_tick_line_color = "#6BCB77"
fig.axis.minor_tick_line_color = "#61A4BC"
fig.axis.axis_label_text_font_size = "20px"
fig.axis.axis_label_text_font_style = "bold"
fig.axis.axis_label_standoff = 12

fig.xaxis.axis_label = "Body Mass"

fig.yaxis.axis_label = "Flipper Length (mm)"
fig.yaxis.major_label_orientation = "vertical"

# adjust axes range
fig.x_range = Range1d(start=2500, end=6500, bounds=(2000, 7000))
fig.y_range = Range1d(start=160, end=240, bounds=(150, 250))

colors = pd.Series(
    ["coral", "aqua", "darkslateblue"], index=["Adelie", "Chinstrap", "Gentoo"]
)
data["colors"] = data["species"].map(colors, na_action="ignore")

# adding glyphs
fig.circle(
    x=data["body_mass_g"],
    y=data["flipper_length_mm"],
    color=data["colors"],
    radius=data["bill_length_mm"],
    radius_dimension="min",
    radius_units="data",
    alpha=0.5,
)

# save and show the figure
show(fig)
```

## use ColumnDataSource to provide data to glyphs

```python
from bokeh.transform import factor_cmap
from bokeh.palettes import Pastel1_4

# create an output file called penguins.html
# output_file("output-files/penguins.html")

# create a figure object
fig = figure()

categorical_species = data["species"].unique().tolist()
categorical_cmap = factor_cmap("species", Pastel1_4, factors=categorical_species)

data_source = ColumnDataSource(data)

# tools
fig.tools = [PanTool(), ResetTool()]
fig.add_tools(HoverTool())
fig.toolbar.logo = None

fig.width = 1000
fig.height = 800

# style the title
fig.title.text = "Penguins mass vs flipper length (in mm)"
fig.title.text_font = "arial"
fig.title.text_font_size = "24px"
fig.title.align = "center"

# style the axes
fig.axis.major_tick_line_color = "#6BCB77"
fig.axis.minor_tick_line_color = "#61A4BC"
fig.axis.axis_label_text_font_size = "20px"
fig.axis.axis_label_text_font_style = "bold"
fig.axis.axis_label_standoff = 12

fig.xaxis.axis_label = "Body Mass"

fig.yaxis.axis_label = "Flipper Length (mm)"
fig.yaxis.major_label_orientation = "vertical"

# adjust axes range
fig.x_range = Range1d(start=2500, end=6500, bounds=(2000, 7000))
fig.y_range = Range1d(start=160, end=240, bounds=(150, 250))

# adding glyphs
fig.circle(
    x="body_mass_g",
    y="flipper_length_mm",
    color=categorical_cmap,
    radius="bill_length_mm",
    radius_dimension="min",
    radius_units="data",
    alpha=0.7,
    source=data_source,
    legend_field="species",
)

fig.legend.location = "top_left"
fig.legend.background_fill_color = "azure"

# save and show the figure
show(fig)
```

```python
from bokeh.sampledata.periodic_table import elements

elements = elements.dropna()
ele_data_source = ColumnDataSource(elements)

standard_state_categories = elements["standard state"].unique().tolist()
state_color_mapper = factor_cmap(
    "standard state", Pastel1_4, factors=standard_state_categories
)

hover_tool = HoverTool()
hover_tool.tooltips = [
    ("Atomic Radius", "@{atomic radius}"),
    ("Boiling " "Point", "@{boiling point}"),
    ("Standard State", "@{standard state}"),
]
fig = figure(plot_width=800, plot_height=600)
fig.tools = [
    PanTool(),
    ResetTool(),
    hover_tool,
]

fig.axis.axis_label_standoff = 10
fig.axis.axis_label_text_font_size = "16px"
fig.axis.axis_label_text_font_style = "italic"
fig.axis.axis_label_text_align = "center"
fig.axis.axis_label_text_font = "arial"

fig.xaxis.axis_label = "Atomic Radius"
fig.yaxis.axis_label = "Boiling Point"

fig.circle(
    x="atomic radius",
    y="boiling point",
    source=ele_data_source,
    radius="van der Waals radius",
    radius_dimension="min",
    color=state_color_mapper,
    alpha=0.7,
    legend_field="standard state",
)

fig.legend.background_fill_color = "azure"

show(fig)
```
